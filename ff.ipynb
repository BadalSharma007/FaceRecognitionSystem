```python
import cv2
import numpy as np
from tf_keras.models import load_model
import pickle

# Load the face recognition model
model = load_model('face_recognition_model.h5')

# Load target names (assuming they were saved during training)
with open('target_names.pkl', 'rb') as f:
    target_names = pickle.load(f)

# Initialize the face detector (Haar Cascade)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Start capturing video from the webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    if not ret:
        print("Failed to grab frame")
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Loop through each detected face
    for (x, y, w, h) in faces:
        # Extract the face region of interest (ROI)
        gray_face = gray[y:y+h, x:x+w]

        # Resize the face to the expected size of the model (37x50)
        face_resized = cv2.resize(gray_face, (37, 50))  # Resize to match the model's expected input size

        # Normalize pixel values
        face_resized = face_resized / 255.0

        # Expand dimensions to fit the model's input shape (None, 50, 37, 1)
        face_resized = np.expand_dims(face_resized, axis=-1)  # Add channel dimension (grayscale: 1 channel)
        face_resized = np.expand_dims(face_resized, axis=0)   # Add batch dimension

        # Make a prediction
        prediction = model.predict(face_resized)
        predicted_class = np.argmax(prediction)

        # Get the predicted class label
        label = target_names[predicted_class]

        # Draw the label and rectangle around the face
        cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Display the resulting frame
    cv2.imshow('Face Recognition', frame)

    # Quit when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
